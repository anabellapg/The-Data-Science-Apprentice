```
def grafica(i=0,j=0):

  flor1=plt.scatter(iris[Y==0,i], iris[Y==0,j], c='red')
  flor2=plt.scatter(iris[Y==1,i], iris[Y==1,j], c='orange')
  flor3=plt.scatter(iris[Y==2,i], iris[Y==2,j], c='green')

  plt.title("Dataset Iris")
  if i==0:
    plt.xlabel("Sepal length")
  if i==1:
    plt.xlabel("Sepal width")
  if i==2:
    plt.xlabel("Petal length")
  if i==3:
    plt.xlabel("Petal width")
  if j==1:
    plt.ylabel("Sepal width")
  if j==2:
    plt.ylabel("Petal length")
  if j==3:
    plt.ylabel("Petal width")

  plt.legend((flor1,flor2,flor3),('Setosa', 'Versicolour', 'Virginica'))
  plt.show()
  return

for i in range(4):
  for j in range(4):
    if i<j:
      grafica(i,j)
```

```
def PredictFlower(flores=np.ones((2,4))):
  filas = flores.shape[0]
  tipos=[]
  for i in range(filas):
    if flores[i,2]<2:
      tipos.append(0)
    elif flores[i,3]<1.6 and flores[i,2]<5.1:
      tipos.append(1)
    else:
      tipos.append(2)
  return tipos
```

```
prediccion=PredictFlower(X[:,:])
aciertos=0

for i in range(Y.shape[0]):
  if Y[i]==prediccion[i]:
    aciertos+=1

tasa=(aciertos/150)*100
print("La tasa de aciertos del modelo es del", tasa, "%.")
```
