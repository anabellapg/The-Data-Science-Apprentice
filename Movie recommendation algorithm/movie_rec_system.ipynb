```python
import pandas as pd
import numpy as np
```

```python
#Cargamos tabla de etiquetas
tags = pd.read_csv('tags.dat', delimiter='\t',header=None)
tags.columns = ['TagID', 'Tag', 'TagPopularity']
del tags['TagPopularity']

#Cargamos tabla de relevancia de las etiquetas
relevance = pd.read_csv('tag_relevance.dat', delimiter='\t',header=None)
relevance.columns = ['MovieID', 'TagID', 'Relevance']

#Cargamos tabla de películas
movies = pd.read_csv('movies_movielens.dat', delimiter='\t',header=None)
movies.columns = ['MovieID', 'Title', 'MoviePopularity']
del movies['MoviePopularity']
```

```python
datos = pd.merge(tags, relevance, on='TagID')
datos = pd.merge(datos, movies, on='MovieID')

datos = datos.drop(columns=['TagID'])
```

```python
datos
```

```python
datos['Year'] = datos['Title'].str.extract('\((\d{4})\)')
datos['Title'] = datos['Title'].str.replace('\(\d{4}\)', '').str.strip()
```

```python
datos
```

```python
datos['Title'] = datos['Title'].str.split(', ').apply(lambda x: ' '.join(reversed(x)))
datos['MovieID'] =pd.factorize(datos['MovieID'])[0]
```

```python
datos
```

```python
datos=datos[datos.Relevance >= 0.5]
column_order = ['MovieID','Title', 'Year', 'Tag','Relevance']

datos = datos[column_order]
datos
```

```python
matrix = datos.pivot(index='MovieID', columns='Tag', values='Relevance')
matrix = matrix.fillna(0)
```

```python
matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similitud_coseno = cosine_similarity(matrix, matrix)
```

```python
similitud_coseno
```

```python
similitud_coseno.shape
```

```python
def content_recommender(title, cosine_sim=similitud_coseno, df=datos):
    # obtenermos el indice
    id_ = datos[datos['Title'] == title]['MovieID'].iloc[0]

    # filtramos por el índice y devolvemos una tupla con el MovieID de
    # otra película y el resultado de la similitud coseno
    scores = list(enumerate(cosine_sim[id_]))

    # ordenamos en base a la similitud coseno
    scores = sorted(scores, key=lambda x: x[1], reverse=True)

    # Nos quedamos con los 10 primeros
    scores = scores[1:7]

    # Extraemos los indices
    ids = [i[0] for i in scores]

    # Seleccionamos los títulos
    titles = datos[datos['MovieID'].isin(ids)]

    # Devolvemos los 6 títulos más similares
    return titles.drop_duplicates('Title')['Title']
```
